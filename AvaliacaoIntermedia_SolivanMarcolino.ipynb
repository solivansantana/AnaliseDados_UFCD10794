# Breve descrição teórica:

A biblioteca Plotly é uma poderosa ferramenta de visualização de dados em Python que permite criar gráficos interativos e de alta qualidade. Ela oferece várias funcionalidades para criar uma ampla variedade de tipos de gráficos, desde gráficos simples até visualizações de dados complexas. Um dos submódulos mais úteis e populares da biblioteca Plotly é o `plotly.express`, que oferece uma interface simplificada para criar gráficos de forma rápida e eficiente. O `plotly.express` foi projetada para simplificar a criação de gráficos. Ela permite que seja possível criar gráficos complexos com apenas algumas linhas de código, economizando tempo e esforço.

# Exemplos Gráficos:

Vamos usar um conjunto de dados fictício para criar os gráficos. Certifique-se de que tenha a biblioteca Plotly instalada e importada antes de executar o código abaixo.

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({
    'Tempo': [1, 2, 3, 4, 5],
    'Valor1': [10, 12, 8, 15, 11],
    'Valor2': [8, 9, 7, 14, 10]
})

gantt = pd.DataFrame({
    'Tarefa': ['Instalação Gitlab', 'Pipeline Gitlab', 'Deployment Aplicação', 'Instalação K8s'],
    'Início': ['2023-09-30', '2023-09-30', '2023-10-01', '2023-10-02'],
    'Fim': ['2023-10-05', '2023-10-03', '2023-10-04', '2023-10-06']
})

maps = pd.DataFrame({
    'Cidade': ['Nova Iorque', 'Paris', 'Toquio', 'Sydney'],
    'Latitude': ['40.7128', '48.8566', '35.682839', '-33.8688'],
    'Longitude': ['-74.0060', '2.3522', '139.759455', '151.2093'],
    'Valor': [10, 15, 8, 12]
})

nuvem = pd.DataFrame({
    'Tempo': [1, 2, 3, 4, 5],
    'Valor1': ['Abacaxi', 'Janela', 'Felicidade', 'Coelho', 'Girasol'],
    'Valor2': ['Computador', 'Lápis', 'Chocolate', 'Montanha', 'Esmeralda']
})

# A. Um Gráfico de linhas:
fig1 = px.line(data, x='Tempo', y=['Valor1', 'Valor2'], title='Gráfico de Linhas')

# B. Um Gráfico de dispersão:
fig2 = px.scatter(data, x='Tempo', y='Valor1', title='Gráfico de Dispersão')

# C. Um Gráfico de barras:
fig3 = px.bar(data, x='Tempo', y='Valor1', title='Gráfico de Barras')

# D. Um Gráfico de setores:
fig4 = px.pie(data, names='Tempo', values='Valor1', title='Gráfico de Setores')

# E. Um Histograma:
fig5 = px.histogram(data, x='Valor1', title='Histograma')

# F. Um Gráfico de bolhas:
fig6 = px.scatter(data, x='Tempo', y='Valor1', size='Valor2', title='Gráfico de Bolhas')

# G. Um Gráfico de Gantt
fig7 = px.timeline(gantt, x_start='Início', x_end='Fim', y='Tarefa')

fig7.update_yaxes(categoryorder="total ascending")
fig7.update_layout(
    title='Gráfico de Gantt',
    xaxis_title='Período',
    yaxis_title='Tarefas',
)

# H. Um Gráfico Boxplot:
fig8 = px.box(data, y='Valor1', title='Gráfico Boxplot')

# I. Um Violin plot
fig9 = px.violin(data, y='Valor1', title='Violin Plot')

# J. Um Gráfico de mapas
fig10 = px.scatter_geo(maps, 
                     lat='Latitude',
                     lon='Longitude',
                     hover_name='Cidade',
                     size='Valor',
                     projection='natural earth')

fig10.update_geos(projection_type="orthographic")

# K. Um Heatmap
heatmap_data = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
fig11 = px.imshow(heatmap_data, title='Heatmap')

# L. Um Gráfico 3D
fig12 = px.scatter_3d(data, x='Tempo', y='Valor1', z='Valor2', title='Gráfico 3D')

# M. Uma Nuvem de palavras
# Requer biblioteca wordcloud adicionalmente. Para instalar a mesma, é necessário fornecer o comando `pip install wordcloud`
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = ' '.join(nuvem['Valor1'].astype(str) + ' ' + nuvem['Valor2'].astype(str))

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nuvem de Palavras')
plt.show()


# Exibir todos os gráficos
fig1.show()
fig2.show()
fig3.show()
fig4.show()
fig5.show()
fig6.show()
fig7.show()
fig8.show()
fig9.show()
fig10.show()
fig11.show()
fig12.show()
```
